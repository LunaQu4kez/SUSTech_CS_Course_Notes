```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
iris = datasets.load_iris()
```

```python
data=pd.DataFrame(iris.data)
```

```python
data.columns=['sepal_length',	'sepal_width',	'petal_length',	'petal_width']
```

```python
data['species']=iris.target
```

```python
iris=data
```

## Matplotlib 

```python
#新建空图
fig  = plt.figure(figsize=(12,10))
#创建子图
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

#画子图1，选择颜色，线型，标签，显示图例
#折线图
ax1.plot(iris['sepal_length'], 'k--', label='sepal_length')
ax1.legend(loc ='best')

#设置绘图种类，以及bin的个数
#直方图
# histgram
a = ax2.hist(iris['sepal_width'], bins =10, label='sepal_width')
ax2.legend(loc='best')
#选择颜色，线型
#散点图
ax3.scatter(range(len(iris['petal_width'])),iris['petal_width'] , label= 'petal_width')
ax3.legend(loc='best')
```

```python
#drawstyle用于选择插值方式
plt.plot(iris['petal_length'], drawstyle = 'steps')
#选择显示范围
plt.plot(iris['petal_length'], drawstyle = 'steps')
plt.xlim([0, 50])
```

```python
fig  = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#设置坐标轴的显示
ax.plot(iris['sepal_length'], 'r-', label = '_nolengend_')
#设置坐标位置
ticks = ax.set_xticks([30, 60 , 90, 120])
#设置该位置显示的文字，以及文字旋转角度，文字大小
labels = ax.set_xticklabels(['30', '60', '90', '120'], rotation = 0, fontsize = 'large')
```

```python
#箱线图
fig  = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#whis指定离群点分割线
ax.boxplot(iris['sepal_length'], whis=2)
#设置标题
plt.title('box plot')
#设置坐标轴名称
plt.ylabel('data content')
fig  = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#whis指定离群点分割线
ax.boxplot(iris['sepal_length'], whis=2)
#设置标题
plt.title('box plot')
#设置坐标轴名称
plt.ylabel('data content')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Fixing random state for reproducibility
np.random.seed(19680801)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(100*np.random.rand(20))

formatter = ticker.FormatStrFormatter('$%1.2f')
ax.yaxis.set_major_formatter(formatter)

for tick in ax.yaxis.get_major_ticks():
    tick.label1On = False
    tick.label2On = True
    tick.label2.set_color('green')

plt.show()
```

# Seaborn

```python
## 介绍seaborn, 散点图
## 导包
import seaborn as sns
sns.set(style='darkgrid')

s1 = sns.jointplot(x="sepal_length", y="petal_length", data = iris)
s1.fig.set_figwidth(8)
s1.fig.set_figheight(6)
```

```python
s2 =sns.FacetGrid(iris, hue="species", height=5).map(plt.scatter, "sepal_length", "sepal_width").add_legend() 
s2.fig.set_figwidth(8)
s2.fig.set_figheight(6)
```

```python
#kind参数
s3 = sns.jointplot(x="sepal_length", y="petal_length", data=iris, kind="reg")
s3.fig.set_figwidth(8)
s3.fig.set_figheight(6)
```

```python
s4 = sns.jointplot(x="sepal_length", y="petal_length", data=iris, kind="hex")
s4.fig.set_figwidth(8)
s4.fig.set_figheight(6)
```

```python
s5 = sns.jointplot(x="sepal_width", y="petal_length", data=iris, kind="kde", space=0, color="b")
s5.fig.set_figwidth(8)
s5.fig.set_figheight(6)
```

```python
## 参数jitter
## 图的叠加
fig = plt.figure(figsize=(8,6))
sns.boxplot(x="species", y="petal_length", data=iris)
sns.stripplot(x="species", y="petal_length", data=iris, jitter=True, edgecolor="gray") 
```
